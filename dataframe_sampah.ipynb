```python
import pandas as pd
```

```python
#nomer 1 menampilkan data dari opendata jabar
df_excel = pd.read_excel("p.xlsx")
df_excel
```

```python
#nomer 2 menghitung data sampah tahunan
jumlah_produksi_sampah_tahun2015 = 0

for a, b in df_excel.iterrows():
    if b["tahun"] == 2015:
        jumlah_produksi_sampah_tahun2015 += b["jumlah_produksi_sampah"]

print(f"jumlah sampah tahun 2015 adalah: {jumlah_produksi_sampah_tahun2015:.1f}")
```

```python
#nomer 3 menghitung jumlah sampah per tahun 2015-2023
total_produksi_sampah_per_tahun = {}

for a, b in df_excel.iterrows():
    tahun = b['tahun']
    total_produksi_sampah_per_tahun[tahun] = total_produksi_sampah_per_tahun.get(tahun, 0) + b['jumlah_produksi_sampah']

total_produksi_sampah_per_tahun_df = pd.DataFrame(list(total_produksi_sampah_per_tahun.items()), columns=['Tahun', 'Total Produksi Sampah'])
total_produksi_sampah_per_tahun_df
```

```python
#nomer 4 menghitung data sampah Kota/Kabupaten per tahun
total_produksi_sampah_per_kota_tahun = {}

for a, b in df_excel.iterrows():
    c = (b['nama_kabupaten_kota'], b['tahun'])
    total_produksi_sampah_per_kota_tahun[c] = total_produksi_sampah_per_kota_tahun.get(c, 0) + b['jumlah_produksi_sampah']

total_produksi_sampah_per_kota_tahun_df = pd.DataFrame([(c[0], c[1], value) for c, value in total_produksi_sampah_per_kota_tahun.items()],columns=['Nama Kabupaten/Kota', 'Tahun', 'Total Produksi Sampah'])
total_produksi_sampah_per_kota_tahun_df
```

```python
total_produksi_sampah_per_kota_tahun_df.to_csv("dataframe sampah.csv", index=False)
```

```python
total_produksi_sampah_per_kota_tahun_df.to_excel("dataframe sampah excel.xlsx", index=False)
```
